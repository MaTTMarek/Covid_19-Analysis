### 7. Answering the questions

##### Where was the most Covid cases?

```python
# looking for countries withe the highest number of total_cases
df.groupby(['location'])['total_cases'].sum().nlargest(20).reset_index()
```

```python
# I am using the location column because country names are easier to read.
# The location column also shows the names of the continents
# that are not included in the classification.
# 1.United States
# 2.India
# 3.Brazil
# 4.France
# 5.United Kingdom
```

#### Despite the gaps in the data, is there a chance to compare the results of the 5 richest countries with the 5 poorest?

```python
# checking how many cases were in most richest countries
df.groupby(['location'])['gdp_per_capita','total_cases'].sum().nlargest(20,columns='gdp_per_capita').reset_index()
```

```python
# checking how many cases were in most poorest countries
df.groupby(['location'])['gdp_per_capita','total_cases'].sum().nsmallest(20,columns='gdp_per_capita').reset_index()
```
